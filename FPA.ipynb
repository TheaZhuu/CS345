```python
# remoave all punctuations
def remove_punctuation(str):
    punc = '''!()-[]{};:'"\,<>./?@#$%^&*_~...'''
    for ele in str:  
        if ele in punc:  
            str = str.replace(ele, "") 
    yield str

comment_nopunc = next(remove_punctuation(comment))
# print(comment_nopunc)

with open("C:\\Users\\12672\\Desktop\\FPA\\comment.txt",'w',encoding="utf-8") as f:
    for line in comment_nopunc:
        f.write(line)
f.close()
```

```python
inputs = open("C:\\Users\\12672\\Desktop\\FPA\\comment.txt",'r',encoding="utf-8")
outputs = open("C:\\Users\\12672\\Desktop\\FPA\\stop_seg_word.txt",'w',encoding="utf-8")
```

```python
# remove stopwords and tokenize the sentence
for line in inputs:
    sentence_depart = word_tokenize(line.strip())
    stopwords = [line.strip() for line in open("C:\\Users\\12672\\Desktop\\FPA\\stopwords.txt",encoding='UTF-8').readlines()]
    temp = ''
    for i in sentence_depart:
        if i not in stopwords:
            if i != '\t':
                temp += i
                temp +=" "
    line_seg = temp
    outputs.writelines(line_seg + '\n')
    
outputs.close()
inputs.close()
```

```python
from sklearn.model_selection import train_test_split

# astype('U'): convert reviews to unicode which passed as dtype 
X_train, X_test, y_train, y_test=train_test_split(data.reviews.values.astype('U'),data.recommend.values,test_size=0.1,random_state=1)
 
print(X_train.shape,X_test.shape)

# print(X_train,X_test)
```

I use TF-IDF and Count vectorizers to convert tanining set and testing set of text to numbers for analysis. It is showing below:

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer

# Convert a collection of raw documents to a matrix of TF-IDF features
tf_vectorizer = TfidfVectorizer()

X_train_tfvec = tf_vectorizer.fit_transform(X_train)
X_test_tfvec = tf_vectorizer.transform(X_test)
print(X_train_tfvec,X_test_tfvec)
print("--------------a matrix of TF-IDF features-----------------")
# Convert a collection of text documents to a matrix of token counts
c_vectorizer = CountVectorizer()

X_train_cvec = c_vectorizer.fit_transform(X_train)
X_test_cvec = c_vectorizer.transform(X_test)
print(X_train_cvec,X_test_cvec)
print("--------------a matrix of token counts-------------------")
```

In this part, I am using K-nearest bneighbors to find in the training data and testing data that is closest to be classified by using TF-IDF and Count vectorizers training set and testing set.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# using TF-IDF features

classifier_tf = KNeighborsClassifier()
classifier_tf.fit(X_train_tfvec,y_train)
y_pred_train = classifier_tf.predict(X_train_tfvec)
train_accuracy = accuracy_score(y_pred_train,y_train)
print("Accuracy for training set by using knn",train_accuracy)


y_pred_test = classifier_tf.predict(X_test_tfvec)
test_accuracy = accuracy_score(y_pred_test,y_test)
print("Accuracy for testing set by using knn",test_accuracy)
```

```python
# using CountVectorizer features 

classifier_c = KNeighborsClassifier()
classifier_c.fit(X_train_cvec,y_train)
y_pred_train = classifier_c.predict(X_train_cvec)
train_accuracy = accuracy_score(y_pred_train,y_train)
print("Accuracy for training set by using knn",train_accuracy)


y_pred_test = classifier_c.predict(X_test_tfvec)
test_accuracy = accuracy_score(y_pred_test,y_test)
print("Accuracy for testing set by using knn",test_accuracy)
```

As the result, I can noticed that the accuracy for both training set and testing set are very high, it is the best performing classifier in knn.
