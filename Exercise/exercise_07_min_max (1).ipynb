* Let's consider the problem of data normalization/scaling:  in statistics and machine learning it is often desirable to convert data to a specified range, for example scale the data so that the values falls between 0 and 1, which is called [min-max rescaling](https://en.wikipedia.org/wiki/Feature_scaling#Rescaling_(min-max_normalization).  
Suppose you have a vector (or one dimensional array) of values `x`, then the min-max rescaled values are given by:
$$
x' = \frac{x-\min(x)}{\max(x) - \min(x)}
$$
